```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Dùng để gọi ra tên các dataset mà Seaborn tích hợp sẵn dùng cho luyện tập
sns.get_dataset_names()
```

```python
#Load dataset có tên 'mpg'
mpg = sns.load_dataset('mpg')
```

```python
mpg
```

```python
#Tính toán tương quan giữa các giá trị numeric trong bảng dữ liệu, có 3 cách chọn: True, False và None.
mpg.corr(numeric_only=True)
```

```python
#Vẽ ma trận tương quan giữa các biến numeric
sns.heatmap(mpg.corr(numeric_only=True))
```

```python
#Vẽ đồ thị scatter plot
sns.scatterplot(data=mpg, x='weight', y='mpg')
```

```python
#Cài đặt hiển thị nền background phía sau của đồ thị
sns.set_theme()
```

```python
#cài đặt style background phía sau của đồ thị, có thể lựa chọn giữa: 'white', 'dark', 'whitegrid', 'darkgrid'
sns.set_style('white')
```

```python
#Cài đặt kích thước tổng thể cho đồ thị, Các yếu tố như kích thước chữ, độ dày đường nét và kích thước điểm sẽ được điều chỉnh dựa trên ngữ cảnh mà bạn chọn.
#Có thể thay đổi tham số với ngữ cảnh khác nhau, Seaborn cung cấp 4 ngữ cảnh chính: 'paper', 'notebook', 'talk' và 'poster'.
sns.set_context('paper', font_scale = 1.2)
```

```python
#Thêm phần màu cho các catagory 
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin')
```

```python
#Thay đổi bảng màu của các categories
#Có thể truy cập vào website của Seaborn để xem các bảng màu khác nhau
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin', palette='mako')
```

```python
#Thay đổi kích thước khung hình của đồ thị
plt.figure(figsize=(10,5))
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin', palette='mako')
```

```python
#Loại bỏ đường viền của đồ thị, có tổng cộng 4 đường viền: top, bottom, left và right. Mặc định sẽ là True (tức là đồng ý loại bỏ)
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin', palette='mako')
sns.despine(left=True)
#Câu lệnh này được sử dụng sau khi vẽ đồ thị, vì theo logic thì cần có đồ thị trước mới xóa đường viền được.
#sns.despine(right=True)
#sns.despine(top=True)
#sns.despine(bottom=True)
#sns.despine(left=True)
#sns.despine(left=False)
#sns.despine(left=True, right=False)
#...
```

```python
#Chuyển đổi dạng đồ thị thành dạng đường
sns.lineplot(data=mpg, x='model_year', y='mpg')
```

```python
#Chuyển đổi đồ thị thành dạng điểm
sns.pointplot(data=mpg, x='model_year', y='mpg')
```

```python
#Vẽ biểu đồ KDE (kernal density estimate - biểu đồ ước lượng mật độ hạt nhân) cho 2 biến
sns.kdeplot(data=mpg, x='weight', y='mpg')
```

```python
#Thêm màu vào biên dưới các đường của biểu đồ KDE cho 2 biến
sns.kdeplot(data=mpg, x='weight', y='mpg', fill=True)
```

```python
#Sắp xếp các biểu đồ theo dạng nằm ngang hoặc nằm dọc
#Đầu tiên, cần chuyển sang dataset 'Tips'
tip = sns.load_dataset('tips')
tip
```

```python
#Dùng FacetGrid để biểu diễn theo chiều dọc
sns.FacetGrid(tip, row= 'day')
```

```python
#Dùng FacetGrid để biểu diễn theo chiều ngang
sns.FacetGrid(tip, col='day')
```

```python
#Dùng FacetGrid để biểu diễn 2 biến categories
sns.FacetGrid(tip, row='day', col='sex')
```

```python
#Dùng FacetGrid để biểu diễn 2 biến categories
sns.FacetGrid(tip, row='sex', col='day')
```

```python
#thêm dữ liệu vào để vẽ histogram
g= sns.FacetGrid(tip, row='sex', col='day')
g.map_dataframe(sns.histplot, "tip")
```

```python
#thêm dữ liệu vào để vẽ scatterplot
g= sns.FacetGrid(tip, row='sex', col='day')
g.map_dataframe(sns.scatterplot, x='tip', y='total_bill')
```
